[Look-and-say sequence](https://en.wikipedia.org/wiki/Look-and-say_sequence)

```python
from utils.general import list_of_chars_to_string
```

```python
def count_look_and_say(curr_str):
    
    curr_char = curr_str[0]
    curr_count = 1
    
    for idx, this_char in enumerate(curr_str[1:]):
        if this_char == curr_char:
            curr_count += 1
        else:
            yield curr_count, curr_char
            curr_count = 1
            curr_char = this_char
            
    yield curr_count, curr_char
    
```

```python
def next_look_and_say_element(curr_str):
    
    counts = [(str(count), char) for count, char in count_look_and_say(curr_str)]
    
    return list_of_chars_to_string([out_el for sublist in counts for out_el in sublist])
```

```python
a = ['1']

for v in range(30):
    next_las = next_look_and_say_element(a[-1])
    print v+1, next_las
    a.append(next_las)
```

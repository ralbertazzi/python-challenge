```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open('data/balloons.jpg')
plt.figure(figsize=(14, 10))
plt.imshow(image)
```

```python
image = np.array(image)
image.shape
```

```python
hw = image.shape[1] // 2
x = image[:, :hw, :] - image[:, hw:, :]

plt.figure(figsize=(14, 10))
plt.imshow(x)
```
